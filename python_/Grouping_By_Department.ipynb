```python
import pandas as pd
import numpy as np
```

Upload Main Data To Notebook

```python
main_data = pd.read_csv('D:\\OneDrive\\Desktop\\PROJECT A\\HR_Dataset_Original\\general_data.csv',
 index_col = "EmployeeID",
 usecols=["EmployeeID",'Age', 'BusinessTravel', 'Department', 'Gender', 'JobLevel', 'JobRole',
       'MaritalStatus', 'MonthlyIncome', 'NumCompaniesWorked',
       'PercentSalaryHike', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager'])
      
```

Replacing Null Values

```python
main_data.interpolate(inplace=True)
```

Defining Average Function

```python
def average(a,b) :
    return a/b
```

Adding Average working time to main data to analyse employee nature

```python
main_data["Avg_Working_Time"]  = average(main_data["TotalWorkingYears"],main_data["NumCompaniesWorked"])
```

Replacing Job Level Number With Job Level Name 

```python
main_data["JobLevel"].replace({1:"Entry Level", 2: "Junior Level", 3: "Mid Level", 4 : "Senior Level",  5: "Executive"}, inplace= True)
```

Separating Each Department's Data Using Group By Method For Further Analysis 

```python
hr_data = main_data.groupby("Department").get_group("Human Resources")
r_d_data = main_data.groupby("Department").get_group("Research & Development")
sale_data = main_data.groupby("Department").get_group("Sales")
```

Uploading Both Employee Survey & Manager Survey Results To Notebook

```python
emp_survey = pd.read_csv('D:\\OneDrive\\Desktop\\PROJECT A\\HR_Dataset_Original\\employee_survey_data.csv', index_col="EmployeeID")
mngr_survey = pd.read_csv('D:\\OneDrive\\Desktop\\PROJECT A\\HR_Dataset_Original\\manager_survey_data.csv', index_col="EmployeeID")
```

Replacing Employee Survey Result Numbers With Relevant Values Using Replace Function

```python
emp_survey["EnvironmentSatisfaction"].replace({1:"Low", 2 : "Medium", 3: "High", 4 : "Very High"}, inplace= True)
emp_survey["JobSatisfaction"].replace({1:"Low", 2 : "Medium", 3: "High", 4 : "Very High"}, inplace= True)
emp_survey["WorkLifeBalance"].replace({1:"Bad", 2 : "Good", 3: "Better", 4 : "Best"}, inplace= True)
emp_survey
```

Replacing Manager Survey Result Numbers With Relevant Values Using Replace Function

```python
mngr_survey["JobInvolvement"].replace({1:"Low", 2 : "Medium", 3: "High", 4 : "Very High"}, inplace= True)
mngr_survey["PerformanceRating"].replace({1:"Low", 2 : "Good", 3: "Excellent", 4 : "Outstanding"}, inplace= True)
mngr_survey
```

Joining Main Data With Survey Data Using Pandas Functions

METHOD 1 (Using Merge Function)

```python
main_empSurvey1 = pd.merge(main_data, emp_survey, how="inner", on = "EmployeeID")
main_empSurvey1
```

```python
main_mngrSurvey1 = pd.merge(main_data,mngr_survey, how = "inner", on = "EmployeeID")
main_mngrSurvey1
```

Method 2 (Using Concat Function)

```python
main_empSurvey2 = pd.concat([main_data, emp_survey], axis=1)
main_empSurvey2
```

```python
main_mngrSurvey2 = pd.concat([main_data, mngr_survey], axis=1)
main_mngrSurvey2
```

Exporting Merged Data As CSV Files Into Python Folder 

```python
main_empSurvey1.to_csv("D:\\OneDrive\\Desktop\\PROJECT A\\python_\\main_empSurvey1.csv")
main_mngrSurvey1.to_csv("D:\\OneDrive\\Desktop\\PROJECT A\\python_\\main_mngrSurvey1.csv")
```
